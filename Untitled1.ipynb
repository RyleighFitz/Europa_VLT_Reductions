```python
#imgs = tel_cor['225']
#comb_mask = masks['225']
#def create_combined_cube(imgs, mask):

def create_combined_cube(imgs, comb_mask):
    combined_cube = np.zeros((imgs[0].data.shape[0],128,128))

    #Limits beyond which data is bad
    min_lambda = 100 #1.457
    max_lambda = 2050 #2.432

    #To Do: Find scale factor for corner images
    #Get indices for sum (excluding 1.8-1.92 microns)
    wave = ru.get_wave(imgs[0])
    idx = (wave>1.92)+(wave<1.8)
    indices = np.where(idx)[0]
    indices = [i for i in indices if i>min_lambda and i<max_lambda]

    pix_scales = {}
    pix_scales[1] = []
    pix_scales[2] = []
    pix_scales[3] = []
    pix_scales[4] = []


    for i in range(128):
        for j in range(128):
            if comb_mask[i,j] >= 6: #both center and corner data exists
            
                cen_sum = np.nansum(imgs[0].data[:,i-32,j-32][indices])
            
                im_num = int(comb_mask[i,j] - 5)
                x_shift, y_shift = ru.get_img_shift(imgs[0].header['CRVAL1'], 
                                          imgs[0].header['CRVAL2'], 
                                          imgs[im_num].header['CRVAL1'], 
                                          imgs[im_num].header['CRVAL2'])

                x_shift, y_shift = int(x_shift), int(y_shift) 
                lx,ly = ru.get_img_loc(x_shift, y_shift)
            
                corn_sum = np.nansum(imgs[im_num].data[:,i-ly,j-lx][indices])
            
                pix_scales[im_num].append(cen_sum/corn_sum)
            
    f = {}
    for key in pix_scales.keys():
        f[key] = np.median(pix_scales[key])

    for k in range(len(imgs[0].data)):
        #Fill in data based on mask
        combined = np.zeros((128,128))
        for j in range(combined.shape[1]):
            for i in range(combined.shape[0]):
                if comb_mask[i,j] == 5: #Use center img
                    combined[i,j] = imgs[0].data[k,i-32,j-32]

                elif comb_mask[i,j]<5 and comb_mask[i,j]>0:
                    im_num = int(comb_mask[i,j])
                    x_shift, y_shift = ru.get_img_shift(imgs[0].header['CRVAL1'], 
                                          imgs[0].header['CRVAL2'], 
                                          imgs[im_num].header['CRVAL1'], 
                                          imgs[im_num].header['CRVAL2'])

                    x_shift, y_shift = int(x_shift), int(y_shift) 
                    lx,ly = ru.get_img_loc(x_shift, y_shift)

                    combined[i,j] = f[im_num]*imgs[im_num].data[k,i-ly,j-lx] #Use corner img
                    #combined[i,j] = 1.1*imgs[im_num].data[k,i-ly,j-lx]

                elif comb_mask[i,j] >= 6: #Use corner and center img
                    cen = imgs[0].data[k,i-32,j-32]
                    im_num = int(comb_mask[i,j] - 5)
                    x_shift, y_shift = ru.get_img_shift(imgs[0].header['CRVAL1'], 
                                          imgs[0].header['CRVAL2'], 
                                          imgs[im_num].header['CRVAL1'], 
                                          imgs[im_num].header['CRVAL2'])

                    x_shift, y_shift = int(x_shift), int(y_shift) 
                    lx,ly = ru.get_img_loc(x_shift, y_shift)
                    ecc = f[im_num]*imgs[im_num].data[k,i-ly,j-lx]
                    #ecc = imgs[im_num].data[k,i-ly,j-lx]

                    avg = np.mean([cen, ecc])
                    combined[i,j] = avg
            
            combined_cube[k] = combined
    
    return combined_cube
```

```python
def get_ortho_coords(img, r):

    #get image coordinates [-1:1, disk centered at (a,b) = (0,0)]
    width, height = 128, 128
    w, h = int(width/2), int(height/2)
    i,j = np.meshgrid(range(128), range(128))
    a = (j-w)/w      #TO DO: Get this to be -1 to 1 (right now it's max(a) is not exactly +1)
    b = (h-i)/h

    a = a/r
    b = b/r
    #Rotate based on NP angle
    A = [[a],[b]]
    alpha = -20 #degrees
    alpha = np.radians(alpha)
    rot_matrix = np.array(( (np.cos(alpha), -np.sin(alpha)),
                        (np.sin(alpha),  np.cos(alpha)) ))
    test = np.einsum('ji, mni -> jmn', rot_matrix, np.dstack([a,b]))
    a_ = test[0]
    b_ = test[1]
    
    return a_, b_
```

```python
@dataclass
class SpatialPatch:
    """Class for keeping track of lat/lon patch."""
    intensity: float
    xcoords: np.array
    "Longitude in orthographic coordinates."
    ycoords: np.array
    "Latitude in orthographic coordinates."
    cen_lat: str
    "Central latitude of measurement. TO DO: maybe add meta data eslewhere"

    def dist_from_center(self) -> float:
        return self.xcoords[0][0]**2 + self.ycoords[0][0]**2
```

```python
from typing import List

def split_into_patches(intensity: np.ma.masked_array, xs: np.ndarray, ys: np.ndarray, cen_lat: str, exclude_masked_data=True) -> List[SpatialPatch]:
    """ Split the data """  
    assert xs.shape[0] == ys.shape[0] == intensity.shape[0] + 1
    assert xs.shape[1] == ys.shape[1] == intensity.shape[1] + 1
    
    patches = []

    for idx, val in np.ndenumerate(intensity):
        l,m = idx
        
        if exclude_masked_data:
            if not intensity.mask[l,m]:
                patches.append(SpatialPatch(
                                    intensity=val,
                                    xcoords=xs[l:l+2, m:m+2],
                                    ycoords=ys[l:l+2, m:m+2],
                                    cen_lat=str(cen_lat)
                                ))
        else:
            patches.append(SpatialPatch(
                            intensity=val,
                            xcoords=xs[l:l+2, m:m+2],
                            ycoords=ys[l:l+2, m:m+2],
                                    cen_lat=str(cen_lat)
                        ))
            
    return patches
```

```python
from copy import deepcopy
def mask_wrapped_pixels(mmx, ortho_coord, cen_lon=0, cen_lat=0):
    
    plate = ccrs.PlateCarree()
    ortho = ccrs.Orthographic(central_longitude=cen_lon, central_latitude=cen_lat)
    earth_r = ortho._x_limits[1]
    transformed = plate.transform_points(ortho, ortho_coord[0]*earth_r, ortho_coord[1]*earth_r)
    lons, lats = transformed[:,:,0], transformed[:,:,1]
    
    lons = np.nan_to_num(lons)
    lats = np.nan_to_num(lats)
    
    dlthresh = 10.0
    for (i,j),_ in np.ndenumerate(mmx[:-1,:-1]):
        xs = [lons[i,j], lons[i,j+1], lons[i+1,j], lons[i+1,j+1]]
        diffs = [np.abs(a-b) for a in xs for b in xs]
        if any(d > dlthresh for d in diffs):
            mmx.mask[i,j] = True
            
    lon_border_mask = np.pad((0 < lons[:-1,:])*(0 > lons[1:,:]), [(0,1),(0,0)])
    
    return lons, lats, mmx*(1-lon_border_mask)
    

```
